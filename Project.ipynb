<h3> Task 1 </h3>

**1. Describe in a few words what is the binomial distribution and what are its parameters and their meaning.**

a Binomial distribution happens when you do $n$ mutually independent experiments. Each of these experiments have the probability of $p$, and we denote the success of each experiment with $1$, and their failiure with $0$. Notation: $X$ ~ $(n, p)$


**2. Plot the binomial distribution with p = 0.7 and size n = 26. Add a suitable
title to your plot.**

```r
# We use ggplot2 to create the plots
library(ggplot2)

p <- 0.7
n <- 26
x <- 0:n

# Create the binomial distribution
BinomDist <- dbinom(x, size = n, prob = p)

# Create the data frame
BinomDataframe <- data.frame(x, BinomDist)

# Create the bar plot using ggplot2
ggplot(data = BinomDataframe, aes(x = x, y = BinomDist)) +
      geom_col(fill = "lightblue", color = "#252525") +
      labs(title = "Binomial Distribution with p = 0.7 and n = 26",
            x = "Number of Successes",
            y = "Probability") +
      theme_minimal() +
      theme(plot.title = element_text(hjust = 0.5, size = 16, face = "bold"), 
            axis.text = element_text(size = 12),
            axis.title = element_text(size = 14),
            plot.title.position = "plot")

options(repr.plot.width = 18, repr.plot.height = 8)
```

**1. Describe in a few words what is the Poisson distribution and what are its parameters and their meaning.**

a Poisson distribution is when some **independent** events happen with the average rate of $λ$ per time unit. $X$ is the number of the events. Notation: $X$ ~ $Poisson(λ)$

**2. Plot the Poisson distribution for $λ$ = 7 and $k$ = 0, 1, . . . , 30.**

```r
lambda <- 7
k <- 0:30

# Create the Poisson distribution
PoissonDist = dpois(k, lambda)

# Create the data frame
PoissonDataframe <- data.frame(k = k, y = PoissonDist)

# Create the bar plot using ggplot2
ggplot(PoissonDataframe, aes(x = k, y = PoissonDist)) +
     geom_col(fill = "lightblue", color = "#252525") +
     labs(title = paste("Poisson Distribution with λ = ", lambda),
          x = "Number of Events", y = "Probability") +
     theme_minimal() +
     theme(plot.title = element_text(hjust = 0.5, size = 16, face = "bold"), 
          axis.text = element_text(size = 12),
          axis.title = element_text(size = 14),
          plot.title.position = "plot")
```

**1. Describe in a few words what is the normal distribution and what are its parameters and their meaning.**

a Normal distribution (Gaussian distribution) is a continuous probability distribution for a random variable ($X$). It has two parameter: The mean ($μ$) and the standard deviation ($σ$).

Mean is the arithmetic mean or average, and standard deviation shows the spread of data.

**2. Plot the distribution for $μ$ = −3 and $σ$ = 0.4 for $x$ ∈ [−5, −1]**

```r
mu <- -3
sigma <- 0.4

# Create a sequence of 100 numbers between -5 and -1
x <- seq(-5, -1, length.out = 100)

# Calculate the normal distribution for each value in x
y <- dnorm(x, mean = mu, sd = sigma)

# Create the plot using ggplot2
ggplot(data.frame(x = x, y = y), aes(x, y)) +
    geom_line(color = "lightblue", linewidth = 2) +
    scale_x_continuous(limits = c(-5, -1)) +
    labs(title = "Normal Distribution with μ = -3, σ = 0.4", x = "x", y = "Density") +
    theme_minimal() +
    theme(plot.title = element_text(hjust = 0.5, size = 16, face = "bold"), 
        axis.text = element_text(size = 12),
        axis.title = element_text(size = 14))
```

<h3> Task 5 </h3>

**A box has 10 coins: six fair ones (prob. of a heads θ = 0.5), two unfair ones (θ = 0.25 and θ = 0.75) and two magical cheating coins (θ = 0 and θ = 1). They all look and feel the same. A coin is taken randomly from the box, and its type Θ has the distribution**

| θ | 0 | 0.25 | 0.5 | 0.75 | 1 |
| --- | --- | --- | --- | --- | --- |
| P(Θ = θ) | 0.1 | 0.1 | 0.6 | 0.1 | 0.1

**Recall the law of total probability and Bayes’ rule and solve the following problems:**

<br>

**1. What is the probability of flipping a heads with the unknown coin?**

P(H) = P(H | θ = 0)P(θ = 0) + P(H | θ = 0.25)P(θ = 0.25) + P(H | θ = 0.5)P(θ = 0.5) + P(H | θ = 0.75)P(θ = 0.75) + P(H | θ = 1)P(θ = 1) = (0 * 0.1) + (0.25 * 0.1) + (0.5 * 0.6) + (0.75 * 0.1) + (1 * 0.1) = 0.5

$$OR$$

```r
Type <- c(0, 0.25, 0.5, 0.75, 1)
Probability <- c(0.1, 0.1, 0.6, 0.1, 0.1)

headProbability <- sum(Probability * Type)

print(paste("Probability of flipping a head: ", headProbability))
```

<br>

**2. You flip the coin and get heads. Call this event H. Conditioned on H, what is the probability that the coin has Θ = 0.25?**

P(Θ = 0.25 | H) = P(H | Θ = 0.25) * P(Θ = 0.25) / P(H) = 0.25 * 0.1 / 0.5 = 0.05

$$OR$$

```r
result = 0.25 * Probability[which(Type == 0.25)] / headProbability

print(paste("Probability of Θ = 0.25 given that the flip was a head: ", result))
```

<br>

**3. Calculate the “posterior probabilities” of P (Θ = θ | H) for θ ∈ {0, 0.25, 0.5, 0.75, 1}. The case θ = 0.25 was solved in item 2.**

```r
PosteriorProb <- list()

for (i in 1:length(Type)) {
    x <- Type[i]
    result <- x * Probability[which(Type == x)] / headProbability
    PosteriorProb[[i]] <- result
    names(PosteriorProb)[[i]] <- as.character(x)  # convert numeric value to character for naming
    print(sprintf("Probability of Θ = %s given that the flip was a head:  %s", x, result))
}
```

<br>

**4. Plot a bar chart of the prior distribution P (Θ = θ) and of the posterior distribution P (Θ = θ | H).**

```r
# create a data frame for plotting
df <- data.frame(Type = as.character(Type), 
                 Probability = Probability,
                 Posterior_Prob = unlist(PosteriorProb))

# plot using ggplot
ggplot(df, aes(x = Type, y = Probability, fill = "Prior")) +
  geom_bar(stat = "identity", alpha = 0.5, width = 0.4) +
  geom_bar(aes(y = Posterior_Prob, fill = "Posterior"), stat = "identity", alpha = 0.5, width = 0.4) +
  scale_fill_manual(name = "Probability", values = c("blue", "red")) +
  labs(title = "Prior and Posterior Distributions",
       subtitle = "Probability of Θ given a head on the flip",
       x = "Type (Θ)", y = "Probability") +
  theme_bw()
```
